# 12. Transportation Network Acquisition <a id='12-transportation-network-acquisition'></a>

This section retrieves the transportation road network for Washington, DC using OpenStreetMap and OSMnx.

Transportation-network acquisition is a foundational step in emergency routing workflows because route optimization and service accessibility analysis depend on a valid navigable road network.

The transportation network will later support:

- shortest-path routing
- emergency response accessibility analysis
- utility service coverage modeling
- operational routing simulation
- network-based spatial analysis

The workflow uses drivable road infrastructure because the project focuses on simulated utility emergency vehicle response operations.

## 12.1 Transportation Network Analysis Context

Unlike Euclidean distance analysis, transportation-network analysis models movement along connected road infrastructure.

This approach is more realistic for emergency-response workflows because travel occurs along road segments rather than straight-line geographic distance.

Transportation-network workflows are commonly used in:

- emergency management
- utility operations
- logistics optimization
- transportation planning
- service-area analysis
- infrastructure accessibility modeling

The road network is retrieved directly from OpenStreetMap using OSMnx to support a reproducible and open-source GIS workflow.

````python
# 12.1: Define transportation network retrieval workflow

transportationNetworkType = "drive"

print("Transportation network type:", transportationNetworkType)
````

````python
# 12.2: Download drivable transportation network

roadNetworkGraph = ox.graph_from_place(
    PLACE_NAME,
    network_type=transportationNetworkType,
    simplify=True
)

print(
    "Transportation network downloaded successfully."
)
````

## 12.3 Transportation Graph Conversion

OSMnx stores transportation networks internally as graph structures composed of nodes and edges.

For GIS analysis and visualization workflows, the transportation graph must be converted into GeoDataFrames representing:

- transportation nodes
- road segments

This conversion enables:

- spatial visualization
- GIS exports
- spatial database integration
- overlay analysis
- geometry processing
- QGIS interoperability

The resulting GeoDataFrames will later support emergency-routing visualization and service-area analysis workflows.

````python
# 12.3: Convert transportation graph into GeoDataFrames

roadNetworkNodesGdf, roadNetworkEdgesGdf = (
    ox.graph_to_gdfs(
        roadNetworkGraph
    )
)
print(f"Transportation graph converted successfully | Nodes: {len(roadNetworkNodesGdf)} | Road Segments: {len(roadNetworkEdgesGdf)}")
````

## 12.4 Transportation Layer Reprojection

The transportation layers are currently stored in a geographic coordinate system.

For reliable distance-based GIS analysis, the workflow reprojects the transportation layers into a projected coordinate reference system.

Projected coordinate systems are required for:

- routing analysis
- service-area modeling
- buffer analysis
- accessibility calculations
- infrastructure distance measurements

The workflow uses:

```text
EPSG:26918 — NAD83 / UTM Zone 18N

````python
# 12.4: Reproject transportation layers

roadNetworkNodesProjectedGdf = (
    roadNetworkNodesGdf.to_crs(PROJECTED_CRS)
)

roadNetworkEdgesProjectedGdf = (
    roadNetworkEdgesGdf.to_crs(PROJECTED_CRS)
)

print(f"Transportation layers reprojected successfully | CRS: {PROJECTED_CRS}")
````

## 12.5 Exploratory Transportation-Network Visualization

- QGIS interoperability

# 15. Emergency Routing Analysis <a id='14-emergency-routing-analysis'></a>

This section performs transportation-network routing analysis between simulated utility depots and outage locations.

Emergency routing analysis is a core operational GIS workflow used in:

- emergency management

The workflow uses NetworkX graph-based routing analysis to estimate optimized transportation paths across the Washington, DC road network.

## Routing Analysis Context

Transportation-network routing differs from Euclidean-distance analysis because movement occurs along connected transportation infrastructure rather than straight-line geographic distance.

Routing analysis commonly incorporates:

- road connectivity
- network topology
- travel accessibility
- operational constraints
- route optimization

The workflow uses shortest-path analysis along the transportation network to estimate realistic emergency-response routes between utility depots and outage locations.

## 15.1 Operational Layer Reprojection for Routing

OSMnx nearest-node operations require geographic coordinate systems using longitude and latitude coordinates.

This section temporarily reprojects the operational GIS layers into WGS84 geographic coordinates to support nearest-node matching within the transportation graph.

The projected layers remain available for distance-based GIS analysis and visualization workflows.

````python
# 15.1: Reproject operational layers into WGS84

utilityDepotsWgs84Gdf = (
    utilityDepotsProjectedGdf.to_crs(
        "EPSG:4326"
    )
)

utilityOutagesWgs84Gdf = (
    utilityOutagesProjectedGdf.to_crs(
        "EPSG:4326"
    )
)

print("Operational layers reprojected into WGS84 successfully.")
````

## 15.2 Transportation Node Matching

Transportation-network routing analysis requires mapping operational locations onto the underlying transportation graph.

This section identifies the nearest transportation-network nodes for:

- utility depots
- outage locations

These nearest-node relationships establish the transportation-network origin and destination points required for shortest-path routing analysis.

````python
# 15.2: Match utility depots to transportation nodes

utilityDepotNodeIds = []

for _, depotRow in utilityDepotsWgs84Gdf.iterrows():

    nearestDepotNode = ox.distance.nearest_nodes(
        roadNetworkGraph,
        X=depotRow.geometry.x,
        Y=depotRow.geometry.y
    )

    utilityDepotNodeIds.append(
        nearestDepotNode
    )

utilityDepotsWgs84Gdf[
    "nearest_node_id"
] = utilityDepotNodeIds

print("Utility depot node matching completed successfully.")
````

````python
# 15.3: Match outage locations to transportation nodes

utilityOutageNodeIds = []

for _, outageRow in utilityOutagesWgs84Gdf.iterrows():

    nearestOutageNode = ox.distance.nearest_nodes(
        roadNetworkGraph,
        X=outageRow.geometry.x,
        Y=outageRow.geometry.y
    )

    utilityOutageNodeIds.append(
        nearestOutageNode
    )

utilityOutagesWgs84Gdf[
    "nearest_node_id"
] = utilityOutageNodeIds

print("Outage node matching completed successfully.")
````

## 15.4 Shortest-Path Emergency Routing Analysis

This section calculates optimized transportation-network routes between utility depots and outage locations.

The workflow uses shortest-path analysis along the transportation graph to estimate operational emergency-response routes.

Each outage location is assigned to the nearest available utility depot using transportation-network accessibility.

The routing analysis estimates:

- operational response paths
- transportation accessibility
- estimated response efficiency
- depot-to-outage connectivity

````python
# 15.4: Perform shortest-path emergency routing analysis

emergencyResponseRoutes = []

for _, outageRow in utilityOutagesWgs84Gdf.iterrows():

    outageNodeId = outageRow[
        "nearest_node_id"
    ]

    bestRoute = None
    bestRouteLength = float("inf")
    bestDepotName = None

    for _, depotRow in utilityDepotsWgs84Gdf.iterrows():

        depotNodeId = depotRow[
            "nearest_node_id"
        ]

        try:

            routeNodeIds = nx.shortest_path(
                roadNetworkGraph,
                source=depotNodeId,
                target=outageNodeId,
                weight="length"
            )

            routeLength = nx.shortest_path_length(
                roadNetworkGraph,
                source=depotNodeId,
                target=outageNodeId,
                weight="length"
            )

            if routeLength < bestRouteLength:

                bestRouteLength = routeLength
                bestRoute = routeNodeIds
                bestDepotName = depotRow[
                    "depot_name"
                ]

        except nx.NetworkXNoPath:

            continue

    if bestRoute:

        routeCoordinates = [

            (
                roadNetworkGraph.nodes[nodeId]["x"],
                roadNetworkGraph.nodes[nodeId]["y"]
            )

            for nodeId in bestRoute
        ]

        emergencyResponseRoutes.append({

            "outage_id": outageRow[
                "outage_id"
            ],

            "assigned_depot": bestDepotName,

            "route_distance_meters": round(
                bestRouteLength,
                2
            ),

            "geometry": LineString(
                routeCoordinates
            )
        })

print("Emergency routing analysis completed successfully.")
````

## 15.5 Emergency Route GeoDataFrame Creation

The emergency-response routing results are converted into a GeoDataFrame to support:

- GIS visualization
- spatial analysis
- QGIS interoperability
- spatial database integration
- web mapping workflows

Each route geometry represents a simulated transportation-network response path between a utility depot and an outage location.

The routing results also preserve operational attributes including:

- assigned utility depot
- outage identifier
- estimated transportation distance

````python
# 15.5: Convert emergency routes into GeoDataFrame

emergencyResponseRoutesGdf = gpd.GeoDataFrame(
    emergencyResponseRoutes,
    crs="EPSG:4326"
)

emergencyResponseRoutesProjectedGdf = (
    emergencyResponseRoutesGdf.to_crs(
        PROJECTED_CRS
    )
)

print("Emergency response route GeoDataFrame created successfully.")
print("Total emergency routes:", len(emergencyResponseRoutesProjectedGdf))
````

## 17.4 Validate GeoPackage Layers with DuckDB SQL

This step validates the exported GeoPackage layers using SQL queries.

The validation query checks the feature count for each operational GIS layer stored in the local spatial database.

This replaces the earlier PostGIS validation approach with a lightweight DuckDB Spatial workflow while preserving the same professional database-validation concept.

````python
# 17.4: Validate GeoPackage layer export using DuckDB SQL

tableCountQuery = f"""
SELECT 'roads' AS layer_name, COUNT(*) AS feature_count
FROM ST_Read('{localSpatialDatabasePath}', layer='roads')

UNION ALL

SELECT 'utility_depots' AS layer_name, COUNT(*) AS feature_count
FROM ST_Read('{localSpatialDatabasePath}', layer='utility_depots')

UNION ALL

SELECT 'utility_outages' AS layer_name, COUNT(*) AS feature_count
FROM ST_Read('{localSpatialDatabasePath}', layer='utility_outages')

UNION ALL

SELECT 'emergency_response_routes' AS layer_name, COUNT(*) AS feature_count
FROM ST_Read('{localSpatialDatabasePath}', layer='emergency_response_routes')

UNION ALL

SELECT 'utility_service_areas' AS layer_name, COUNT(*) AS feature_count
FROM ST_Read('{localSpatialDatabasePath}', layer='utility_service_areas')

UNION ALL

SELECT 'outage_impact_zones' AS layer_name, COUNT(*) AS feature_count
FROM ST_Read('{localSpatialDatabasePath}', layer='outage_impact_zones')

UNION ALL

SELECT 'roads_near_outages' AS layer_name, COUNT(*) AS feature_count
FROM ST_Read('{localSpatialDatabasePath}', layer='roads_near_outages');
"""

spatialLayerCountsDf = duckdbConnection.execute(
    tableCountQuery
).df()

spatialLayerCountsDf
````

## 17.5 Emergency Route Metrics Using Spatial SQL

This section performs SQL-based analysis of the emergency-response routing results.

The analysis evaluates:

- route distances
- depot assignments
- operational accessibility
- routing efficiency

The workflow demonstrates how spatial SQL can support operational GIS analytics directly against local spatial data layers.

````python
# 17.5: Analyze emergency response routes using SQL

emergencyRouteMetricsQuery = f"""
SELECT
    outage_id,
    assigned_depot,
    ROUND(route_distance_meters / 1000.0, 2)
        AS route_distance_km
FROM ST_Read(
    '{localSpatialDatabasePath}',
    layer='emergency_response_routes'
)
ORDER BY route_distance_km DESC;
"""

emergencyRouteMetricsDf = duckdbConnection.execute(
    emergencyRouteMetricsQuery
).df()

emergencyRouteMetricsDf
````

## 17.6 Average Emergency Response Distance

This section calculates the average transportation-network response distance across all simulated outage events.

Average route-distance analysis is commonly used in operational GIS systems to evaluate:

- emergency accessibility
- operational efficiency
- infrastructure coverage performance
- dispatch optimization conditions

````python
# 17.6: Calculate average emergency response distance

averageResponseDistanceQuery = f"""
SELECT
    ROUND(
        AVG(route_distance_meters) / 1000.0,
        2
    ) AS average_route_distance_km
FROM ST_Read(
    '{localSpatialDatabasePath}',
    layer='emergency_response_routes'
);
"""

averageResponseDistanceDf = duckdbConnection.execute(
    averageResponseDistanceQuery
).df()

averageResponseDistanceDf
````

## 17.7 Spatial SQL Accessibility Analysis

This section evaluates whether simulated outage locations fall inside utility emergency-response service areas.

The workflow uses spatial SQL intersection analysis to identify:

- accessible outage events
- operational service coverage
- infrastructure reachability
- response accessibility conditions

This type of spatial analysis is commonly used in operational GIS systems for:

- infrastructure planning
- dispatch optimization
- emergency-response assessment
- accessibility modeling

````python
# 17.7: Identify outages located inside service areas

outageAccessibilityQuery = f"""
SELECT
    o.outage_id,
    o.priority_level,
    s.service_area_type
FROM ST_Read(
        '{localSpatialDatabasePath}',
        layer='utility_outages'
     ) AS o

JOIN ST_Read(
        '{localSpatialDatabasePath}',
        layer='utility_service_areas'
     ) AS s

ON ST_Intersects(
       o.geom,
       s.geom
   )

ORDER BY o.outage_id;
"""

outageAccessibilityDf = duckdbConnection.execute(
    outageAccessibilityQuery
).df()

outageAccessibilityDf
````

````python
# 19.2: Create Leaflet web map HTML

leafletHtml = """
<!DOCTYPE html>
<html>
<head>
    <meta charset="utf-8">
    <title>Utility Emergency Response Routing Analysis</title>
    <meta name="viewport" content="width=device-width, initial-scale=1.0">

    <link rel="stylesheet" href="https://unpkg.com/leaflet@1.9.4/dist/leaflet.css" />

    <style>
        html, body {
            height: 100%;
            margin: 0;
            font-family: Arial, sans-serif;
        }

        #map {
            width: 100%;
            height: 100%;
        }

        .map-title {
            position: absolute;
            top: 15px;
            left: 50px;
            z-index: 1000;
            background: white;
            padding: 12px 16px;
            border-radius: 6px;
            box-shadow: 0 2px 8px rgba(0,0,0,0.25);
            max-width: 420px;
        }

        .map-title h2 {
            margin: 0 0 6px 0;
            font-size: 18px;
        }

        .map-title p {
            margin: 0;
            font-size: 13px;
            color: #555;
            line-height: 1.4;
        }

        .legend {
            position: absolute;
            bottom: 25px;
            left: 25px;
            z-index: 1000;
            background: white;
            padding: 12px 14px;
            border-radius: 6px;
            box-shadow: 0 2px 8px rgba(0,0,0,0.25);
            font-size: 13px;
            line-height: 1.6;
        }

        .legend-title {
            font-weight: bold;
            margin-bottom: 6px;
        }

        .legend-symbol {
            display: inline-block;
            width: 14px;
            height: 14px;
            margin-right: 6px;
            vertical-align: middle;
        }
    </style>
</head>

<body>

<div id="map"></div>

<div class="map-title">
    <h2>Utility Emergency Response Routing Analysis</h2>
    <p>
        Interactive GIS map showing simulated utility depots, outage events,
        emergency response routes, and service coverage areas in Washington, DC.
    </p>
</div>

<div class="legend">
    <div class="legend-title">Map Legend</div>
    <div><span class="legend-symbol" style="background:#0D47A1;"></span>Utility Depot</div>
    <div><span class="legend-symbol" style="background:#E65100; border-radius:50%;"></span>Outage Event</div>
    <div><span class="legend-symbol" style="background:#1565C0;"></span>Response Route</div>
    <div><span class="legend-symbol" style="background:#90CAF9;"></span>Service Area</div>
    <div><span class="legend-symbol" style="background:#BDBDBD;"></span>Road Network</div>
</div>

<script src="https://unpkg.com/leaflet@1.9.4/dist/leaflet.js"></script>

<script>
    const map = L.map('map').setView([38.9072, -77.0369], 11);

    const baseMap = L.tileLayer(
        'https://{s}.tile.openstreetmap.org/{z}/{x}/{y}.png',
        {
            maxZoom: 19,
            attribution: '&copy; OpenStreetMap contributors'
        }
    ).addTo(map);

    const roadsLayer = L.geoJSON(null, {
        style: {
            color: '#BDBDBD',
            weight: 0.6,
            opacity: 0.55
        }
    });

    const serviceAreasLayer = L.geoJSON(null, {
        style: {
            color: '#1565C0',
            weight: 2,
            fillColor: '#90CAF9',
            fillOpacity: 0.25
        }
    });

    const routesLayer = L.geoJSON(null, {
        style: {
            color: '#1565C0',
            weight: 4,
            opacity: 0.9
        },
        onEachFeature: function(feature, layer) {
            const props = feature.properties;
            layer.bindPopup(
                '<b>Emergency Response Route</b><br>' +
                'Outage ID: ' + props.outage_id + '<br>' +
                'Assigned Depot: ' + props.assigned_depot + '<br>' +
                'Route Distance: ' + Math.round(props.route_distance_meters / 100) / 10 + ' km'
            );
        }
    });

    const depotsLayer = L.geoJSON(null, {
        pointToLayer: function(feature, latlng) {
            return L.circleMarker(latlng, {
                radius: 8,
                fillColor: '#0D47A1',
                color: 'white',
                weight: 1,
                fillOpacity: 0.95
            });
        },
        onEachFeature: function(feature, layer) {
            const props = feature.properties;
            layer.bindPopup(
                '<b>Utility Depot</b><br>' +
                'Depot ID: ' + props.depot_id + '<br>' +
                'Depot Name: ' + props.depot_name
            );
        }
    });

    const outagesLayer = L.geoJSON(null, {
        pointToLayer: function(feature, latlng) {
            return L.circleMarker(latlng, {
                radius: 7,
                fillColor: '#E65100',
                color: 'white',
                weight: 1,
                opacity: 1,
                fillOpacity: 0.95
            });
        },
        onEachFeature: function(feature, layer) {
            const props = feature.properties;
            layer.bindPopup(
                '<b>Utility Outage</b><br>' +
                'Outage ID: ' + props.outage_id + '<br>' +
                'Priority: ' + props.priority_level + '<br>' +
                'Affected Customers: ' + props.affected_customers
            );
        }
    });

    Promise.all([
        fetch('data/roads.geojson').then(r => r.json()),
        fetch('data/utility_service_areas.geojson').then(r => r.json()),
        fetch('data/emergency_response_routes.geojson').then(r => r.json()),
        fetch('data/utility_depots.geojson').then(r => r.json()),
        fetch('data/utility_outages.geojson').then(r => r.json())
    ])
    .then(([roadsData, serviceAreasData, routesData, depotsData, outagesData]) => {

        roadsLayer.addData(roadsData);
        serviceAreasLayer.addData(serviceAreasData);
        routesLayer.addData(routesData);
        depotsLayer.addData(depotsData);
        outagesLayer.addData(outagesData);

        serviceAreasLayer.addTo(map);
        roadsLayer.addTo(map);
        routesLayer.addTo(map);
        depotsLayer.addTo(map);
        outagesLayer.addTo(map);

        const operationalGroup = L.featureGroup([
            roadsLayer,
            serviceAreasLayer,
            routesLayer,
            depotsLayer,
            outagesLayer
        ]);

        map.fitBounds(
            operationalGroup.getBounds(),
            {
                padding: [40, 40]
            }
        );
    });

    const overlayMaps = {
        "Service Coverage Areas": serviceAreasLayer,
        "Road Network": roadsLayer,
        "Emergency Response Routes": routesLayer,
        "Utility Depots": depotsLayer,
        "Outage Events": outagesLayer
    };

    L.control.layers(
        {"OpenStreetMap": baseMap},
        overlayMaps,
        {collapsed: false}
    ).addTo(map);
</script>

</body>
</html>
"""

leafletMapPath.write_text(
    leafletHtml,
    encoding="utf-8"
)

print("Leaflet web map created successfully.")
````

# 21. Limitations <a id='21-limitations'></a>

Although the workflow demonstrates realistic GIS engineering concepts, several analytical simplifications were intentionally used within the project.

The emergency-response routes were calculated using transportation-network distance rather than estimated travel time. Real operational systems would typically incorporate:

- traffic conditions
- road restrictions
- travel-speed assumptions
- dynamic transportation conditions

The service-area analysis also used Euclidean-distance buffer zones rather than true network-based service-area modeling. As a result, the generated coverage areas represent simplified approximations of operational accessibility.

The utility depots and outage events were simulated operational locations and do not represent actual infrastructure systems or real outage conditions.

The workflow additionally relied on OpenStreetMap transportation data, which may contain incomplete or generalized transportation-network information depending on geographic location.

Despite these simplifications, the project successfully demonstrates the core GIS engineering concepts involved in transportation-network routing, operational accessibility analysis, spatial SQL workflows, and web GIS delivery.
